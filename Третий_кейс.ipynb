### **Посчитайте Retention 15 дня (в процентах) для пользователей, зарегистрированных в январе**

Cохраните результат в переменную `retention_15_day`

**Примечание:** результат округлите до 5 знаков после запятой

```python
# Ваше решение
import os, csv
import datetime as dt
r_15 = []
e_15_double=[]
# Получаем список списков пользователей зарегистрированных в январе
with open('registrations.csv', "r", encoding = 'utf-8') as reg:
  rd = csv.reader(reg, delimiter=';')
  for i, elem in enumerate(rd):
    if i==0:
      continue
    if elem[1].split('-')[1]=='01':
      r_15.append(elem)
# Записываем общее количество пользователей зарегистрированных в январе
cnt_user = len(r_15)
# Получаем список списков пользователей заходивших на платформу
with open('entries.csv', "r", encoding = 'utf-8') as entr:
  rd = csv.reader(entr, delimiter=';')
  for i, elem in enumerate(rd):
    if i==0:
      continue
    e_15_double.append(elem)
#Удаляем дубликаты пользователей заходивших на платформу несколько раз в день
e_15 = []
[e_15.append(i)
for i in e_15_double if i not in e_15]
# Сравниваем списки
ret = 0
for r, reg_d in enumerate(r_15):
# Преобразуем строку в datetime
  reg_date = dt.datetime.strptime(reg_d[1], '%Y-%m-%d')
  for e, entr_d in enumerate(e_15):
# Преобразуем строку в datetime
    entr_date = dt.datetime.strptime(entr_d[1], '%Y-%m-%d')
# Получаем deltatime
    td_delta = entr_date - reg_date
# Счетчик кол-ва пользователей заходивших на платформу на 15 день
    if reg_d[0]==entr_d[0] and td_delta.days==15:
      ret +=1
retention_15_day = round(100*ret/cnt_user, 5)
```

```python
#@title ✏️ Проверка: чтобы проверить свое решение запустите код в этой ячейке
# Открываем файл с правильными ответами
with open('metrics.txt', 'r') as f:
    answers = f.read().split('\n')

correct_answer = float(answers[0])

try:
    assert retention_15_day == correct_answer
except AssertionError:
    print('Ответы не совпадают')
else:
    print('Поздравляем, Вы справились!')
```
